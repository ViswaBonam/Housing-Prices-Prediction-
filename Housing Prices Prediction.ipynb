```python
#Defining a funtion to deal with the missing  values in the data
def missing(df):
    for col in df.columns :
        if df[col].dtype=='object':
            df[col]=df[col].replace(np.nan,"None")
        else:
            df[col]=df[col].replace(np.nan, df[col].mean())
```

```python
#The categorical features in the data have a value of 'None' where data isnt available and '0' for numerical features
```

```python
#Feature Engineering
features_selected=['MSSubClass','MSZoning','LotFrontage','LotArea','Street','Neighborhood','OverallQual','YearBuilt','YearRemodAdd','ExterQual','BsmtCond','TotalBsmtSF','Heating','HeatingQC','GrLivArea','FullBath','TotRmsAbvGrd','GarageCars','GarageArea']
train_df[features_selected]
```

```python
#import necessary libraries to handle categorical data
from sklearn.preprocessing import LabelEncoder 

le = LabelEncoder()

for i in features_selected:                    
    if train_df[i].dtype == "object":
        train_df[i] = le.fit_transform(train_df[i])

for i in features_selected:                    
    if test_df[i].dtype == "object":
        test_df[i] = le.fit_transform(test_df[i])
```

```python
Clean_test_data = test_df[features_selected]
```

```python
final_train=train_df[features_selected]
final_train
```

```python
plt.figure(figsize= (18,10))
sns.heatmap(final_train.corr(), annot=True)
```

```python
#Training the data
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
x=final_train
y=train_data['SalePrice']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=1)
```

```python
Linear_reg=LinearRegression()
```

```python
Linear_reg.fit(x_train,y_train)
```

```python
pred=Linear_reg.predict(x_test)
```

```python
mae=mean_absolute_error(pred, y_test)
mae
```

```python
# defining a function to calculate the accuracy

def accuracy(mod):
    mod.fit(x_train,y_train)
    pred=mod.predict(x_test)
    mse = mean_squared_error(y_test, pred)
    print(f'MSE  = {mse}')
    print(f"RMSE = {np.sqrt(mse)}")
```

```python
#building models
```

```python
accuracy(Linear_reg)
```

```python
DecisionTree=DecisionTreeRegressor()
accuracy(DecisionTree)
```

```python
RandomForest=RandomForestRegressor()
accuracy(RandomForest)
```

```python
#AS WE HAVE MOST ACCURACY FROM RANDOM FOREST
RandomForest=RandomForestRegressor()
RandomForest.fit(x_train,y_train)
random_pred= RandomForest.predict(Clean_test_data)
```
